```python
import crf
import matplotlib.pyplot as plt
```

```python
image = crf.imageio.imread('data/2007_000033_0img.png')
label = crf.imageio.imread('data/2007_000033_5labels.png')
#실험을 위해 정답을 통해서 낮은 품질의 세그멘테이션을 만들어냄
unary = crf.synthetic.augment_label(label, num_classes=21)
```

```python
pred, c_pred = crf.CRF(image, unary)
```

```python
plt.subplot(1,2,2)
plt.imshow(pred[1])
plt.title('output')
plt.axis('off')
plt.subplot(1,2,1)
plt.imshow(unary[:,:,1])
plt.title('input')
plt.axis('off')
plt.show()
```
